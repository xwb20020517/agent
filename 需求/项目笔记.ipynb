# 1.访问、刷新令牌

## 访问令牌
    http是无状态的，每次请求（比如查看指定用户的会话内容），后端都需要校验是否登录，所以创建一个登录令牌，后端验证得到用户信息才能修改
## 刷新令牌与缓存机制
    访问令牌一般设置过期时间短，因为经常在前后端传输，容易泄露造成用户信息修改等。为了让登录令牌过期自动刷新，设置refresh令牌，存入缓存是因为经常验证，缓存到时间可以自动删除。
## JWT鉴权
    access_token的创建是包含用户id等内容的，只需使用jwt.encode()即可创建；
    只需使用jwt.decode()即可解析当前用户信息
